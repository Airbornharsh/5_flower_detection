```python
imgs = []
label = []

resized_list_directory = os.listdir(resized_images)

for filefolder in resized_list_directory:
    list_image = os.listdir(resized_images + filefolder)
    for file in list_image:
        img = Image.open(resized_images + filefolder + "/" + file)
        img = np.array(img)/255.0
        # imgs.append(img)
        imgs.append(img)
        if filefolder == "Lilly":
            label.append(0)
        elif filefolder == "Lotus":
            label.append(1)
        elif filefolder == "Orchid":
            label.append(2)
        elif filefolder == "Sunflower":
            label.append(3)
        elif filefolder == "Tulip":
            label.append(4)
        print(filefolder + "/" + file)
        del(img)
        del(file)
    del(filefolder)
```

```python
imgs = np.asarray(imgs)
label = np.asarray(label)
```

```python
# print(imgs.shape)
# print(label.shape)
```

```python
np.unique(label)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(imgs, label, test_size=0.2, random_state=42)
```

```python
del(imgs)
del(label)
```

```python
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

```python
mobilenet_model = "https://tfhub.dev/google/tf2-preview/mobilenet_v2/feature_vector/4"

pretrained_model = hub.KerasLayer(mobilenet_model,input_shape=(224,224,3),trainable=False)
```

```python
model = tf.keras.Sequential([
    pretrained_model,
    tf.keras.layers.Dense(5)
])

model.summary()
```

```python
model.compile(optimizer="adam",loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),metrics=['accuracy'])
```

```python
model.fit(X_train,Y_train,epochs=5)
```

```python
del(X_train)
del(Y_train)
```

```python
score, accuracy = model.evaluate(X_test,Y_test)
```

```python
model.save("5_Flower_Detection.h5")
```
